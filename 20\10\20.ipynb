```python
import psycopg2 as p
import getpass
import pandas as pd
import sqlalchemy
import json
```

```python
USERNAME=input("Enter user name: ")
HOST="rc1c-uja07w9pu5nubjm2.mdb.yandexcloud.net"
DBNAME="taximipt"
PORT="6432"
READ_WRITE = True
password = getpass.getpass(prompt="Enter password for {}: ".format(USERNAME))
```

```python
_attrs = {
    "host": HOST,
    "database": DBNAME,
    "user": USERNAME,
    "password": password,
    #"sslmode": "verify-full",
    "port": PORT,
    "extra": ""
}
if READ_WRITE:
    _attrs["extra"] = "target_session_attrs=read-write"
conn_string = "postgresql://{user}:{password}@{host}:{port}/{database}?{extra}".format(**_attrs)

def conn():
    return sqlalchemy.create_engine(conn_string)
```

```python
data = pd.read_sql("select * from common.offers_dmshem limit 100", conn())
s = data["doc"]
a = list(s)
a = list(map(lambda x : json.loads(x.replace("'", "\"")), a))
print(type(a[0]))
data = pd.DataFrame(a)
data
```
